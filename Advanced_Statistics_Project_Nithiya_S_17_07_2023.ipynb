3.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

```python
mu = 5
sigma = 1.5
# find the cumulative probability
# norm.cdf() calculates the cumulative probability
from scipy.stats import norm
prob = norm.cdf(3.17, mu, sigma)
print(
    "The proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm is",
    round(prob,4)*100,"%"
)
```

3.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?

```python
mu = 5
sigma = 1.5
# find the cumulative probability
# norm.cdf() calculates the cumulative probability
from scipy.stats import norm
prob1 = norm.cdf(3.6, mu, sigma)
prob2 = 1- prob1
print(
    "The proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm is",
    round(prob2,4)*100,"%"
)
```

3.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
mu = 5
sigma = 1.5
# find the cumulative probability
# norm.cdf() calculates the cumulative probability
from scipy.stats import norm
prob1 = norm.cdf(5.5, mu, sigma)
prob2 = norm.cdf(5, mu, sigma)
prob3 = prob1- prob2
print(
    "The proportion of the gunny bags have a breaking strength between 5 and 5.5 per sq cm is",
    round(prob3,4)*100,"%"
)
```

What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
mu = 5
sigma = 1.5
# find the cumulative probability
# norm.cdf() calculates the cumulative probability
from scipy.stats import norm
prob1 = norm.cdf(3, mu, sigma)
prob2 = norm.cdf(7.5, mu, sigma)
prob3 = 1- (prob2 - prob1)
print(
    "The proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm is",
    round(prob3,4)*100,"%"
)
```

```python
mu = 77
sigma = 8.5
# find the cumulative probability
# norm.cdf() calculates the cumulative probability
from scipy.stats import norm
prob_less_than_85 = norm.cdf(85, mu, sigma)
print(
    "The probability that a student will score less than 85 is",
    round(prob_less_than_85,4)
)
```

```python
prob_less_than_87 = norm.cdf(87, mu, sigma)
prob_less_than_65 = norm.cdf(65, mu, sigma)
prob_bet_87_65 = prob_less_than_87 - prob_less_than_65
print(
    "The probability that a student will score between 87 and 65 is",
    round(prob_bet_87_65,4)
)
```

The program is considered successful if the candidate is able to do more than 5 push-ups, as compared to when he/she enrolled in the program. Using the sample data provided can you conclude whether the program is successful? (Consider the level of Significance as 5%)
Note that this is a problem of the paired-t-test. Since the claim is that the training will make a difference of more than 5, the null and alternative hypotheses must be formed accordingly.


```python
# Get the Data

from google.colab import drive
drive.mount('/content/drive' , force_remount=True)
path="/content/drive/MyDrive/Colab Notebooks/Aquarius_gym.csv"
aquarius_data=pd.read_csv(path)
```

```python
aquarius_data = aquarius_data.assign(Diff = aquarius_data.After - aquarius_data.Before)

aquarius_data
```

```python
aquarius_data.describe()
```

The null hypothesis formed is

H0:μ>5

against the alternative hypothesis

Ha:μ<= 5

n=100

```python
# import the required function
from scipy.stats import norm
from scipy.stats import t

# set the values of sample mean and sigma
x_bar, sigma = 5.55, 2.87


# set the value of sample size and degrees of freedom
n = 100
k = n - 1

# construct the confidence interval
np.round(t.interval(0.95, df=k, loc=x_bar, scale=sigma / np.sqrt(n)), 2)
```

Population mean is between 4.98 and 6.12 with 95% confidence

```python
import statistics
sigma = statistics.pstdev(aquarius_data['Diff'], x_bar)

mu = 4.98
# calculate the test statistic
test_stat = (x_bar - mu) / (sigma/np.sqrt(100))
print(test_stat)
```

```python
# calculate the p-value using the cdf() function
1 - norm.cdf(test_stat)
```

As the p-value ~0.023 is lesser than level of significance, we can reject the null hypothesis.
